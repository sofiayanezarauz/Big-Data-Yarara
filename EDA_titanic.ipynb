# Exploratory Data Analysis in Python (Titanic)


Cuando usas un punto después de una variable, es como si estuvieras diciendo: "Quiero abrir este bolsillo específico y ver qué hay adentro o usarlo". Así es como accedes a las herramientas especiales que esa variable tiene para ti.

Por ejemplo, cuando escribes ` data.head()`, es como si estuvieras diciendo: "Quiero usar la herramienta que me muestra las primeras filas".

Imaginemos que tu variable en Python es un robot.

**Atributos:**

Son las características o propiedades del robot. Por ejemplo, su color, su tamaño o la cantidad de batería que le queda.
No realizan acciones por sí mismos. Son como etiquetas que te dicen algo sobre el robot.
Accedes a un atributo sin usar paréntesis. Por ejemplo: `mi_robot.color`


**Métodos:**

Son las acciones o cosas que el robot puede hacer. Por ejemplo, caminar, hablar o encender una luz.
Para que el robot realice una acción, tienes que "llamar" al método, y esto generalmente se hace con paréntesis. Por ejemplo: `mi_robot.caminar()`
Dentro de los paréntesis, a veces debes darle instrucciones adicionales (parámetros) sobre cómo quieres que realice la acción. Por ejemplo: mi_robot.caminar(pasos=5) le diría al robot que camine 5 pasos.


El método `astype()` en pandas te permite cambiar el tipo de datos (o dtype) de una columna.



*   `int`: Convierte los datos a números enteros. Por ejemplo, 5.6 se convertiría en 5.
*   `float`: Convierte los datos a números de punto flotante (es decir, números decimales).

*   `str`: Convierte los datos a cadenas de texto.
*   `bool`: Convierte los datos a valores booleanos (True o False). Los valores 0, NaN, y cadenas vacías suelen convertirse en False, y otros valores en True.

Para usar astype(), simplemente selecciona la columna (o columnas) que deseas convertir y aplica el método con el nuevo tipo deseado.

## 3. Detección y tratamiento de datos ausentes

Si pensamos nuestros datos como un libro, algunas páginas pueden estar en blanco. Es importante identificar estas páginas y decidir qué hacer con ellas.



```python
# Detectamos datos ausentes
data.isnull().sum()
```

```python
# Agregar una nueva fila con valores ausentes
data.loc[len(data)] = None

# Verificar las últimas filas
data.tail()
```

```python
# Tratamiento: por ejemplo, llenar los valores ausentes de la columna "Age" con la mediana
data['Age'].fillna(data['Age'].median(), inplace=True)
```

**Tratamiento de datos ausente**

**Eliminación:**

`dropna():` Esta es la forma más simple. Simplemente eliminas las filas o columnas con datos ausentes. Sin embargo, no siempre es la mejor opción, ya que podrías perder mucha información.

**Imputación:**

`fillna()`: Rellena valores NA/NaN usando el método especificado

1.   *Valor Fijo:* Reemplaza los valores faltantes con un valor fijo, como 0 o cualquier otro valor que tenga sentido en el contexto de tus datos.
2.   *Media, Mediana, Moda:* Para datos numéricos, puedes reemplazar valores faltantes con la media (mean()) o la mediana (median()) de la columna. Para datos categóricos, puedes usar la moda (mode()).
3.  *Interpolación:* Utiliza métodos de interpolación para rellenar valores faltantes, especialmente útil para series temporales. Puedes usar el método interpolate().
4.  *Modelos Predictivos:* Usa modelos de machine learning para predecir y reemplazar los valores faltantes basados en otras columnas. Por ejemplo, puedes entrenar un modelo de regresión para predecir un valor faltante basado en otras características.

## 5. Correlación de variables

La correlación es como ver si dos amigos se llevan bien. Si cuando uno se ríe, el otro también se ríe, ¡probablemente sean buenos amigos!

Así es como encontramos relaciones entre columnas.

```python
# Calculamos y mostramos la correlación entre las variables (Solo mostramos columnas numericas)
correlation_matrix = data.select_dtypes(include=['number']).corr()
sns.heatmap(correlation_matrix, annot=True)
plt.show()
```

###1: ¿Cómo podrías crear una nueva columna que indique si un pasajero era niño, adulto o anciano, basándote en su edad?

```python
#Ejercicio 1

# Función para clasificar por edad
def clasificar_edad(edad):
    if pd.isnull(edad):
        return 'desconocido'
    elif edad < 13:
        return 'niño'
    elif edad < 60:
        return 'adulto'
    else:
        return 'anciano'

# Crear nueva columna con la clasificación
data['Grupo_Edad'] = data['Age'].apply(clasificar_edad)

# Mostrar algunas filas para verificar
print(data[['Age', 'Grupo_Edad']].head(30))
```

```python
#Ejercicio 3
first_class_survivors = data[(data['Pclass'] == 1) & (data['Survived'] == 1)]
first_class_survivors
```

###4: ¿Cómo agruparías a los pasajeros por sexo y clase, y calcularías la tasa de supervivencia promedio para cada grupo?

###5: ¿Cuál es la edad promedio de los pasajeros que sobrevivieron y de los que no?

###6: ¿Existe una correlación significativa entre la edad y la tarifa del pasaje?

```python
#Ejercicio 6
# Filtrar filas donde ambas columnas tengan valores válidos
subset = data[['Age', 'Fare']].dropna()

# Calcular la correlación
correlacion = subset['Age'].corr(subset['Fare'])

print(f"Correlación entre edad y tarifa: {correlacion:.3f}")
```
